```python
import numpy as np

def bigger(list_ex):
  big = []
  for i in range(len(list_ex)):
    try:
      if list_ex[i+1] > list_ex[i]:
        big.append(list_ex[i+1])
    except IndexError:
      break
  return big
def biggest(list_ex):
  big_er = bigger(list_ex)
  while len(big_er) != 1:
    big_er = bigger(big_er)
  return big_er[0]
```
