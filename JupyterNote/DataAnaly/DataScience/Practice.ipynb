```python
def get_counts(sequen):
    counts = {}
    for x in sequen:
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1
    return counts

from collections import defaultdict
def get_counts2(sequen):
    counts_2 = defaultdict(int)
    for x in sequen:
        counts_2[x] += 1
    return counts_2
```

```python
counts_2 = get_counts2(time_zones)
counts_2.clear()
```

```python
#Top 10 time zones
def top_counts(count_dict, n = 10):
    value_key_paris = [(tz, counts) for counts, tz in count_dict.items()]
    value_key_paris.sort(reverse = True)
    return value_key_paris[:n+1]
```

```python
top_counts(counts)
```

```python
by_tz_os = cframe.groupby(['tz', 'os'])
```

```python
agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts
```

```python
indexer = agg_counts.sum(1).argsort()
indexer
```

```python
count_subset = agg_counts.take(indexer[-10:])
count_subset
```

```python
indexer_2 = agg_counts.sum(1).nlargest(10)
indexer_2
```

```python
count_subset = count_subset.stack()
count_subset.name = 'total'
count_subset = count_subset.reset_index()
count_subset
```

```python
import seaborn as sns
sns.barplot(x = 'total', y = 'tz', hue = 'os', data = count_subset)
```

```python
#normalize
def norm_total(group):
    group['normed_total'] = group.total / group.total.sum()
    return group

results = count_subset.groupby('tz').apply(norm_total)
sns.barplot(x = 'normed_total', y = 'tz', hue = 'os', data = results)
```

```python
#Easier way
import seaborn as sns
g = count_subset.groupby('tz')
result2 = count_subset.total / g.total.transform('sum')
count_subset['Normed_total'] = result2
sns.barplot(x = 'Normed_total', y = 'tz', hue = 'os', data = count_subset)
```

```python
import pandas as pd
from collections import deque
years = range(1880, 2011)
pieces = deque()
columns = ['name', 'sex', 'births']

for year in years:
    path = 'https://raw.githubusercontent.com/wesm/pydata-book/2nd-edition/datasets/babynames/yob%d.txt' % year
    frame = pd.read_csv(path, names = columns)
    frame['year'] = year
    pieces.append(frame)

pieces
```

```python
#Concatenate everything into a single DataFrame
names = pd.concat(pieces, ignore_index = True)
names.info()
```

```python
total_births = names.pivot_table('births', index = 'year',
                                 columns = 'sex', aggfunc = sum)
total_births
```

```python
total_births.plot(title = 'Total births by sex and year')
```

```python
def add_prop(group):
    group['prop'] = group.births/group.births.sum()
    return group
names = names.groupby(['year', 'sex']).apply(add_prop)

import seaborn as sns
sns.barplot(x = 'prop', y = 'name', data = names[:20])
```

```python
names.groupby(['year', 'sex']).prop.sum()
```

```python
def get_top1000(group):
    return group.sort_values(by = 'births', ascending = False)[:1000]
top1000 = names.groupby(['year', 'sex']).apply(get_top1000)
top1000
```

```python
top1000.reset_index(inplace = True, drop = True)
top1000
```

```python
pieces = []
for year, group in names.groupby(['year', 'sex']):
    pieces.append(group.sort_values(by = 'births', ascending = False)[:1000])
top1000 = pd.concat(pieces, ignore_index = True)
top1000
```

```python
boys = top1000[top1000.sex == 'M']
girls  =top1000[top1000.sex == 'F']
boys
```

```python
total_births = top1000.pivot_table('births', index = 'year',
                                   columns = 'name', aggfunc = sum)
total_births = total_births.fillna(0)
total_births
```

```python
subset = total_births[['John', 'Harry', 'Mary', 'Marilyn']]
subset.plot(subplots = True, figsize = (12,10), grid = False,
           title = 'Number of births per yaer')
```
