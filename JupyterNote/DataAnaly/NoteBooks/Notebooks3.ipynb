```python


import math
def nCk(n,k):
    Combi = math.factorial(n)/(math.factorial(n-k)*math.factorial(k))
    return Combi
```

```python
n = 100
c = 1
k = c + 1
probability_r = 0.1
probability_c = 1 - probability_r
Bern = 0
for i in range(k, n +1):
    Bern += nCk(n,i)*(probability_r**i)*(probability_c**(n-i))
Bern
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
def func_SOFI(x,a,b,c):
    import numpy as np
    return a*np.exp(-b*x) +c

x_SOFI = np.linspace(0, len(df_SOFI), len(df_SOFI))
y_SOFI = np.reshape(df_SOFI.to_numpy(),len(df_SOFI))
popt, *_ = curve_fit(func_SOFI, x_SOFI, y_SOFI)
popt

y_fit_SOFI = [popt[0]*np.exp(-popt[1]*x) + popt[2] for x in x_SOFI]
plt.plot(x_SOFI,y_SOFI, 'o')
plt.plot(x_SOFI, y_fit_SOFI, '-')
plt.show()
```
